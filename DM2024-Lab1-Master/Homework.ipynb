```python
# note this takes time to compute. You may want to reduce the amount of terms you want to compute frequencies for
term_frequencies = []
for j in range(0,X_counts.shape[1]):
    term_frequencies.append(sum(X_counts[:,j].toarray()))

#[3, 8, 5, 2, 5, 8, 2, 5, 3, 2]
```

```python
term_frequencies = np.asarray(X_counts.sum(axis=0))[0]
```

```python
term_frequencies[0] #sum of first term: 00
```

```python
plt.subplots(figsize=(100, 10))
g = sns.barplot(x=count_vect.get_feature_names_out()[:300], 
            y=term_frequencies[:300])
g.set_xticklabels(count_vect.get_feature_names_out()[:300], rotation = 90);
```

```python
top_n = 50  # Choose the top 50 frequency vocabularies 

# find the indices of vocabularies
top_indices = np.argpartition(term_frequencies, -top_n)[-top_n:]
# fetch the vocabularies and frequencies
top_terms = count_vect.get_feature_names_out()[top_indices]
top_frequencies = term_frequencies[top_indices]

# sort the vocabularies alphabetically
sorted_indices = np.argsort(top_terms)
top_terms_sorted = top_terms[sorted_indices]
top_frequencies_sorted = top_frequencies[sorted_indices]

g = sns.barplot(x=top_terms_sorted, y=top_frequencies_sorted)
g.set_xticklabels(top_terms_sorted, rotation=90)
plt.show()
```

```python
# Answer here
top_n = 50  # Choose the top 50 frequency vocabularies 

# find the indices of vocabularies
top_indices = np.argpartition(term_frequencies, -top_n)[-top_n:]

# fetch the vocabularies and frequencies
top_terms = count_vect.get_feature_names_out()[top_indices]
top_frequencies = term_frequencies[top_indices]

# sort the vocabularies based on frequencies (from high to low)
sorted_indices = np.argsort(top_frequencies)[::-1]  # 按頻率由高到低排序
top_terms_sorted = top_terms[sorted_indices]
top_frequencies_sorted = top_frequencies[sorted_indices]

g = sns.barplot(x=top_terms_sorted, y=top_frequencies_sorted)
g.set_xticklabels(top_terms_sorted, rotation=90)
plt.show()
```
